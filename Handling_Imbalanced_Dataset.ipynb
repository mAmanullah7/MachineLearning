**Imbalanced Dataset:**

A datset with an equal class distribution



```python
# importing the dependecies
import numpy as np
import pandas as pd
```

```python
# laoding the data set to pandas DataFrame
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
# ditribtion of the two classes
credit_card_data['Class'].value_counts()
```

This is a highly imbalanced Dataset

0 --> Legit Transaction

1 --> Fraudulent Transaction

```python
# seperating the legit and fradulent transaction
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-Sampling

Build a simple dataset conataining similar distribution of Legit & Fradulent Transaction

```python
legit_sample = legit.sample(n=3)
```

```python
print(legit_sample.shape)
```

Concatenate the two data frames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)    #axis = 0 means column wise addn
```
